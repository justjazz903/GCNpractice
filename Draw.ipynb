```python
import networkx as nx
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
file = pd.read_csv('preprocessed_data.txt', sep=' ', header = None)
graph_seg_head = 0
for n,graph_seg_end in enumerate(file[file[0] == 'x'].index):
    print(n,graph_seg_head)
    graph_seg_head = graph_seg_end + 1
```

```python
pd_g = file.iloc[122:151]
for i in pd_g[pd_g[0]=='e'][2]:
    print(i)
```

```python
g = nx.Graph()
g.add_node(1,feature='day')
g.add_node(2,feature='water')
g.add_node(3,feature='yesterday')
g.add_node(4,feature='laugh')
g.add_node(5,feature='fight')
g.add_node(6,feature='fun')
g.add_node(7,feature='school')
g.add_node(8,feature='bit')
g.add_node(9,feature='birthday')
```

```python
for i in pd_g[pd_g[0]=='e'][1]:
    g.add_edge(int(i.split(',')[0]),int(i.split(',')[1]))
```

```python
w = []
for i in pd_g[pd_g[0]=='e'][2]:
    w.append(float(i))
w = np.round(np.dot(w,10000),decimals=1)

for n,i in enumerate(g.edges):
    g[i[0]][i[1]]['weight'] = w[n]
pos=nx.get_node_attributes(g,'pos')
e_labels = nx.get_edge_attributes(g,'weight')
```

```python
nx.draw(g,pos=nx.circular_layout(g),labels=dict(g.nodes('feature')),node_size=1000)
nx.draw_networkx_edge_labels(g,pos=nx.circular_layout(g),edge_labels=e_labels)
plt.pyplot.savefig('dataExample.jpg')
```
